````{admonition} Exercice
Que vaut `l` après avoir exécuté le code suivant?
```ocaml
let l = [1; 2; 3] in
0::l
```

```{admonition} Solution
:class: tip, dropdown
`l` vaut toujours `[1; 2; 3]` : écrire `e::l` ne modifie pas `l`.
```
````
